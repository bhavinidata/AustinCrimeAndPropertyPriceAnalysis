```python
import pandas as pd
import csv
import os
import numpy as np
import requests
from config import zip_api_key
```

```python
csv_path = os.path.join("Data", "zip_list.csv")
```

```python
with open(csv_path,"r",newline="\n") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=",")
    
    ## Build List of Zip and County
    zip_ = []
    county = []
    
    for row in csv_reader:
        zip_.append(row[0])
        county.append(row[1])
        
    
```

```python
## Build a Panda DataFrame
initial_dict = {"ZIP": zip_,
                 "COUNTY": county}
```

```python
zip_detail_df = pd.DataFrame(initial_dict)
```

```python
zip_detail_df.head()
```

```python
zip_detail_df["Total Population"] = np.nan
zip_detail_df["HouseHolds"] = np.nan
zip_detail_df["WhitePop"] = np.nan
zip_detail_df["BlackPop"] = np.nan
zip_detail_df["HispanicPop"] = np.nan
zip_detail_df["AsianPop"] = np.nan
zip_detail_df["IndianPop"] = np.nan
zip_detail_df["HawaiianPop"] = np.nan
zip_detail_df["OtherPop"] = np.nan
zip_detail_df["MalePop"] = np.nan
zip_detail_df["FemalePop"] = np.nan
zip_detail_df["PersonsPerHousehold"] = np.nan
zip_detail_df["AverageHouseValue"] = np.nan
zip_detail_df["IncomePerHousehold"] = np.nan
zip_detail_df["MedianAge"] = np.nan
zip_detail_df["MedianAgeMale"] = np.nan
zip_detail_df["MedianAgeFemale"] = np.nan
zip_detail_df["AverageFamilySize"] = np.nan
zip_detail_df["Latitude"] = np.nan
zip_detail_df["Longitude"] = np.nan
```

```python
zip_detail_df.head(1)
```

```python
## Define RAW URL which will be used to build BASE URL dynamically, as zip code is part of the URL
params = {
    "key": zip_api_key
}
RAW_URL="https://api.zip-codes.com/ZipCodesAPI.svc/1.0/GetZipCodeDetails/"
```

```python
## Read zip codes from dataFrame -> Build Base URL -> Make API call -> Populate DataFrame
for index, row in zip_detail_df.iterrows():
    
    zipcode = row["ZIP"]
    BASE_URL = f"{RAW_URL}{zipcode}"
    
    try:
        
        response = requests.get(BASE_URL, params=params)
        data = response.json()
        
        zip_detail_df.loc[index, "Total Population"] = data["item"]["ZipCodePopulation"]
        zip_detail_df.loc[index, "HouseHolds"] = data["item"]["HouseholdsPerZipcode"]
        zip_detail_df.loc[index, "WhitePop"] = data["item"]["WhitePop"]
        zip_detail_df.loc[index, "BlackPop"] = data["item"]["BlackPop"]
        zip_detail_df.loc[index, "HispanicPop"] = data["item"]["HispanicPop"]
        zip_detail_df.loc[index, "AsianPop"] = data["item"]["AsianPop"]
        zip_detail_df.loc[index, "IndianPop"] = data["item"]["IndianPop"]
        zip_detail_df.loc[index, "HawaiianPop"] = data["item"]["HawaiianPop"]
        zip_detail_df.loc[index, "OtherPop"] = data["item"]["OtherPop"]
        zip_detail_df.loc[index, "MalePop"] = data["item"]["MalePop"]
        zip_detail_df.loc[index, "FemalePop"] = data["item"]["FemalePop"]
        zip_detail_df.loc[index, "PersonsPerHousehold"] = data["item"]["PersonsPerHousehold"]
        zip_detail_df.loc[index, "AverageHouseValue"] = data["item"]["AverageHouseValue"]
        zip_detail_df.loc[index, "IncomePerHousehold"] = data["item"]["IncomePerHousehold"]
        zip_detail_df.loc[index, "MedianAge"] = data["item"]["MedianAge"]
        zip_detail_df.loc[index, "MedianAgeMale"] = data["item"]["MedianAgeMale"]
        zip_detail_df.loc[index, "MedianAgeFemale"] = data["item"]["MedianAgeFemale"]
        zip_detail_df.loc[index, "AverageFamilySize"] = data["item"]["AverageFamilySize"]
        zip_detail_df.loc[index, "Latitude"] = data["item"]["Latitude"]
        zip_detail_df.loc[index, "Longitude"] = data["item"]["Longitude"]
        
        
        
        
    except(KeyError, IndexError):
        print("Zip Details not found... skipping.")
    
    
```
