```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
austin_pop_df = pd.read_csv("Data/AustinZipCodePopulation.csv")
austin_pop_df.head(3)
```

```python
population = []
```

```python
population.append(austin_pop_df["2015_Population"].sum())
population.append(austin_pop_df["2016_Population"].sum())
population.append(austin_pop_df["2017_Population"].sum())
population.append(austin_pop_df["2018_Population"].sum())
```

```python
years = ["2015_Austin", "2016_Austin", "2017_Austin", "2018_Austin"]
x_axis = np.arange(len(years))
```

```python
plt.bar(x_axis, population, color="b", align="center")
```

```python
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, years)
```

```python
plt.savefig("Figures/Austin_Population_Bar_Chart")
```

```python
plt.show()
```

```python
two_year_df = austin_pop_df[["2015_Population", "2018_Population"]]
```

```python
multi_plot = two_year_df.plot(kind="bar", figsize=(10,5))
```

```python
multi_plot.set_xticklabels(austin_pop_df["Zipcode"], rotation = 90)
```
